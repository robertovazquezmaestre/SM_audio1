Se importan las librerías necesarias para tabajar con audio y con graficas en python.

```python

from scipy.io import wavfile
import IPython
import os
import numpy as np
import matplotlib.pyplot as plt
```

Se especifican las rutas de los directorios de entrada donde estan los audios con los que se van a trabajar y los de salida donde se guardaran los audios generados.

```python
cwd = os.getcwd()
audio_input_path = os.path.join(cwd, os.path.join('Audio_input'))  
audio_output_path = os.path.join(cwd, os.path.join('Audio_output')) 
print(f'Directorio con los audios de entrada: {audio_input_path}')
print(f'Directorio donde guardaremos los audios generados: {audio_output_path}\n')
```

Se carga el archivo .wav de audio con el que se va a trabajar, en este caso se va a trabbajar con el archivo gamne_of_thrones.wav. También se muestra la fracuencia de muestreo del archivo.

```python
filename = os.path.join(audio_input_path, 'game_of_thrones.wav')
sample_rate, audio_data = wavfile.read(filename)
print(f'Frecuencia de muestreo (sample rate): {sample_rate/1000} kHz')
```

Con este widget se puede escuchar el audio en estéreo.

```python
IPython.display.Audio(audio_data.T, rate=sample_rate) 
```

Aquí se muestran el tamaño del archivo de audio en estéreo. Como el tamaño del audio y el número de canales que tiene, en este caso son dos porque es estéreo.

```python

print('El tamaño del archivo de audio en estereo es:')
!ls -sh Audio_input/game_of_thrones.wav

print('Datos de audio (estereo):')
print(f'- Tamaño:     {audio_data.shape}') 
print(f'- 1º canal:   {audio_data[:5, 0]}...')
print(f'- 2º canal:   {audio_data[:5, 1]}...')
print(f'- Resolucion: {type(audio_data[0,0])}\n')
```

Aquí se cambia el audio de estereo a mono, usando el método de calcular la media.

```python
new_data_mono = audio_data.mean(axis=1)  
print('Nuevos datos de audio (mono):')
print(f'- Nuevo tamaño: {new_data_mono.shape}')
print(f'- Canal unico:  {new_data_mono[:5]}...')

new_data_mono = new_data_mono.astype(np.int16)
print(f'- Resolucion:   {type(new_data_mono[0])}\n')
```

Se guarda el nuvo audio mono en un archivo en la ruta de directori de salida. La tasa de frecuencia del audio mono es la misma que para el estereo.

```python
wavfile.write(
    filename=os.path.join(audio_output_path, 'game_of_thrones_mono.wav'),
    rate=sample_rate,
    data=new_data_mono
)
```

Se muesta la misma información que se ha mostrado para el audio estereo para el mono. Como es de espera el archivo pesa aproximadamente al mitad y tiene un único canal.

```python
print('El tamaño del archivo de audio en mono es:')
!ls -sh Audio_output/game_of_thrones_mono.wav

print('Datos de audio (mono):')
print(f'- Tamaño:     {new_data_mono.shape}') 
print(f'- Canal unico:  {new_data_mono[:5]}...')
print(f'- Resolucion: {type(new_data_mono[0])}\n')
```

Con este widget se puede escuchar el audio mono.

```python
IPython.display.Audio(new_data_mono, rate=sample_rate)
```

Con MatPlotLib se muestra el dominio del tiempo para, primero, el audio estéreo y después el audio mono.

```python

tiempo = np.arange(len(audio_data)) / sample_rate

plt.figure(figsize=(12, 5))

plt.plot(tiempo, audio_data[:, 0], label="Canal Izquierdo")
plt.plot(tiempo, audio_data[:, 1], label="Canal Derecho", alpha=0.7)

plt.title("Señal de Audio en el Dominio del Tiempo (Estéreo)")
plt.xlabel("Tiempo [segundos]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid()
plt.show()
```

```python
tiempo = np.arange(len(audio_data)) / sample_rate

plt.figure(figsize=(10, 4))
plt.plot(tiempo, new_data_mono)
plt.title("Señal de Audio en el Dominio del Tiempo (Mono)")
plt.xlabel("Tiempo [segundos]")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

## Diferencia entre audio estéreo y mono

El audio **mono** utiliza un solo canal de sonido.
Eso significa que toda la información sonora se mezcla en una única señal y se reproduce igual en todos los altavoces o audífonos. No hay sensación de dirección o espacio: el sonido parece venir del mismo punto.

En cambio, el audio **estéreo** utiliza dos canales independientes: uno izquierdo y uno derecho.
Cada canal puede contener información diferente, lo que permite crear una sensación de profundidad, dirección y espacio.
